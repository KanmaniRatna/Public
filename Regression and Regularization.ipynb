- R-squared: The R-squared value is 0.585, indicating that approximately 58.5% of the variability in the "price" can be explained by the included independent variables. This measure represents the proportion of the dependent variable's variance that is accounted for by the independent variables in the model.

Significance of Coefficients:
Coefficients: The coefficients represent the estimated impact of each independent variable on the dependent variable ("price"). For example, for every one-unit increase in the "bedrooms" variable, the model predicts a decrease in price of approximately $34,910.
Standard Errors: The standard errors estimate the variability in the coefficient estimates. Smaller standard errors suggest higher precision in the coefficient estimates.the standard error of 1.88e+04 suggests that the coefficient estimate for the "waterfront" variable is subject to a relatively high amount of uncertainty. The coefficient estimate may vary by approximately ±18,800 units from the true population value.
t-values: The t-values measure the statistical significance of the coefficients. Larger absolute t-values (farther from zero) indicate more statistically significant coefficients.
P-values: The p-values (P>|t|) represent the probability of observing a t-value as extreme as the one obtained if the null hypothesis (no effect of the independent variable on the dependent variable) were true. Smaller p-values (below a significance level, often 0.05) suggest statistical significance.
F-statistic: The F-statistic tests the overall significance of the regression model. In this case, the F-statistic is 3804, and the associated probability  is close to zero, indicating that the model has statistical significance.
Log-Likelihood: The log-likelihood value (-2.9812e+05) represents the maximized log-likelihood for the model, which is used in model comparison.
AIC and BIC: The AIC (Akaike Information Criterion) and BIC (Bayesian Information Criterion) are measures of model fit that penalize models with more parameters. Lower AIC and BIC values (both are 5.963e+05 in this case) indicate better model fit.

```python
# separate features and target
categorical_features = ['bedrooms','bathrooms','floors', 'waterfront','view','condition','grade']
numerical_features = ['sqft_living', 'sqft_above','lat','long','sqft_living15']
target = 'price'
```

```python
house[['bedrooms','bathrooms','floors', 'waterfront','view','condition','grade','sqft_living', 'sqft_above','lat','long','sqft_living15']]
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler
from sklearn.linear_model import LinearRegression, ElasticNet, Ridge, Lasso
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
# split data in train / test
X_train, X_test, y_train, y_test = train_test_split(house[categorical_features + numerical_features]
                                                    ,house[target],
                                                    test_size = 0.3, random_state=1234)
```

```python
# define transformers
cat_transformer = Pipeline(steps = [('onehot', OneHotEncoder(handle_unknown='ignore'))])
num_transformer = Pipeline(steps = [('scaler', StandardScaler())])
```

```python
# building processor
preprocessor = ColumnTransformer(transformers = [('cat',cat_transformer, categorical_features),
                                                 ('num', num_transformer, numerical_features)])
```

```python
linear_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
linear_regression.fit(X_train, y_train)
```

```python
pred = linear_regression.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(pred,y_test))
```

```python
ridge_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', Ridge(alpha=3.0))
])
```

```python
ridge_regression.fit(X_train, y_train)
```

```python
pred_ridge = ridge_regression.predict(X_test)
```

```python
np.sqrt(mean_squared_error(pred_ridge,y_test))
```

```python
lasso_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', Lasso(alpha=1.0))
])
```

```python
lasso_regression.fit(X_train, y_train)
```

```python
lasso_pred = lasso_regression.predict(X_test)
```

```python
np.sqrt(mean_squared_error(lasso_pred, y_test))
```

```python
elasticnet_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', ElasticNet(alpha=1.0, l1_ratio=0.8))
])
```

```python
elasticnet_regression.fit(X_train, y_train)
```

```python
elasticnet_pred = elasticnet_regression.predict(X_test)
```

```python
np.sqrt(mean_squared_error(elasticnet_pred, y_test))
```

Based on the residual analysis on price and sqft . It is noticed that the model is overfitting . Overfitting occurs when the predicted value is greater than the actual vlaue. This exhibits Heteroscedasticity,that is the  variability of errors or residuals is not constant across all levels of the independent variables. it means that the spread or dispersion of the residuals is not the same for all predicted values.

 This violates one of the assumptions of  linear regression, which assumes that the variance of the errors is constant (homoscedasticity). This assumption is important because it affects the efficiency and reliability of the regression estimates, including the coefficient estimates and their standard errors.If the spread increases the difference between PV and AV increases.Model will deviate more.The model is concentrated in the centre.
