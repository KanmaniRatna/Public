3) Describe your data cleaning approach. How and why did you decide on which cleaning techniques to use?
The data cleaning techniques used were removing outliers and Missing value imputations.
 
The outlier in price column was removed as it was an extreme datapoint compared to the given price range. 
There were values identified in units sold boxplot but were not dropped as it is important to use them for analysis.
There were missing values present in the given dataset, as a first step threshold was calculated based on the length of the dataset. 
Since the threshold value was higher than 5% no rows were dropped. 
The missing values of price and units sold were filled by using the impute by subgroup method. They were filled by grouping the mean of each product  group with its respective price range . The missing values in units sold were also filled using the same method of grouping product and unit sold range. 
The column names Sale date and Units Sold were renamed to Sale_date and Units_sold for easy reference.

```python
ds['Product'] = ds['Product'].fillna("Other_Prod")
```

```python
ds['Region'] = ds['Region'].fillna("Other_Reg")
```

```python
ds["Price"] = ds.groupby("Product")["Price"].transform(lambda x: x.fillna(x.mean()))
```

```python
ds["Units_sold"] = ds.groupby("Product")["Units_sold"].transform(lambda x: x.fillna(x.mean()))
```

```python
ds.rename(columns={'Units Sold': 'Units_sold'}, inplace=True)
```

```python
ds.rename(columns={'Sale Date': 'Sale_date'}, inplace=True)
```

```python
ds['Product']=ds['Product'].str.replace('Product','PD')
ds['Product']
```

```python
#Sales per product
grouped_data = ds.groupby('Product')['Units_sold'].sum().reset_index()
print(grouped_data)
```

```python
ds['Units_sold'].sum()
```

```python
fig, ax = plt.subplots(figsize=(40,10))

grouped_data.sort_values('Units_sold', ascending=False, inplace=True)
grouped_data['Units_sold'] = grouped_data['Units_sold'].round()

# To plot the chart, without the filled in Other_Prod value 
excluded_value = 'Other_Prod'  
filtered_data = grouped_data[grouped_data['Product'] != excluded_value]
ax.bar(filtered_data['Product'], filtered_data['Units_sold'])

#labelling the chart
ax.set_xlabel('Product')
ax.set_ylabel('Sum of Units')
ax.set_title('Sum of Units sold in each Product')
ax.bar_label(ax.containers[0])

#to rotate xticks for better readability
plt.xticks(rotation=90)


plt.show()
```
